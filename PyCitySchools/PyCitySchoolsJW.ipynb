```python
# Dependencies and Setup
import pandas as pd

# Load File
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
```

```python
school_df = pd.read_csv(school_data)
print(school_df.shape)
print(school_df.info)
```

```python
student_df = pd.read_csv(student_data)
print(student_df.shape)
print(student_df.info)
```

```python
# combine to create a master df
df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
df.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_df["school_name"])
school_count
```

```python
# Calculate the total number of students
student_count = len(df["student_name"])
print(type(student_count))
student_count
```

```python
# Calculate the total budget
total_budget = school_df["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = df["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = df["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
math_pass = (df["math_score"] >= 70).sum()
math_percentage = (math_pass / student_count) * 100
math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
reading_pass = (df["reading_score"] >= 70).sum()
reading_percentage = (reading_pass / student_count) * 100
reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# passed math and reading 
passed_both = ((df["math_score"] >= 70) & (df["reading_score"] >= 70)).sum()
passed_both_percentage = (passed_both / student_count) * 100
passed_both_percentage
```

```python
# list out the data needed for the summary
data = {"School Count" : school_count, 
        "Student Count" : student_count, 
        "Total Budget" : total_budget,
        "Average Math Score" : average_math_score,
        "Average Reading Score" : average_reading_score,
        "% Passing Math" : math_percentage,
        "% Passing Reading" : reading_percentage,
        "% Passing Math & Reading" : passed_both_percentage}
data
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame(data, [0])

# Formatting
district_summary["Student Count"] = district_summary["Student Count"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the type per school from school_data
school_types = school_df.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count per school from school_data
per_school_counts = df.groupby("school_name")["student_name"].count()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school from school_data
per_school_budget = df.groupby("school_name")["budget"].mean()
print(per_school_budget)

per_school_capita = per_school_budget / per_school_counts
per_school_capita
```

```python
# Calculate the average test scores per school from school_data_complete
per_school_math = df.groupby("school_name")["math_score"].mean()
per_school_reading = df.groupby("school_name")["reading_score"].mean()

print(per_school_math)
per_school_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete
students_passing_math = df[df["math_score"] >= 70].groupby("school_name")["student_name"].count()
students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete
students_passing_reading = df[df["reading_score"] >= 70].groupby("school_name")["student_name"].count()
students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
passing_both_school = df[(df["math_score"] >= 70) & (df["reading_score"] >= 70)].groupby("school_name")["student_name"].count()
passing_both_school
```

```python
# Use the provided code to calculate the passing rates
per_school_math = (students_passing_math / per_school_counts) * 100
per_school_reading = (students_passing_reading / per_school_counts) * 100
passing_rate = (passing_both_school / per_school_counts) * 100

df_rates = pd.DataFrame({
    "Per School Math (%)": per_school_math,
    "Per School Reading (%)": per_school_reading,
    "Passing Rate (%)": passing_rate})
df_rates
```

```python
# make sure to grab all the correct data points
school = {"School Type" : school_types, 
        "Student Count" : per_school_counts, 
        "School Budget" : per_school_budget,
        "Per Student Budget" : per_school_capita,
        "Average Math Score" : per_school_math,
        "Average Reading Score" : per_school_reading,
        "Per School Math (%)": per_school_math,
        "Per School Reading (%)": per_school_reading,
        "Passing Rate (%)": passing_rate}
school
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame(school)

# copy for later before formatting
spending_df = per_school_summary.copy()
```

```python
# Formatting
per_school_summary["School Budget"] = per_school_summary["School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Math Scores by Grade

```python
# boolean mask by grade
ninth_graders = df[(df["grade"] == "9th")]
tenth_graders = df[(df["grade"] == "10th")]
eleventh_graders = df[(df["grade"] == "11th")]
twelfth_graders = df[(df["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_math = ninth_graders.groupby("school_name")["math_score"].mean()
tenth_grade_math = tenth_graders.groupby("school_name")["math_score"].mean()
eleventh_grade_math = eleventh_graders.groupby("school_name")["math_score"].mean()
twelfth_grade_math = twelfth_graders.groupby("school_name")["math_score"].mean()

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
math_scores_by_grade = pd.DataFrame ({ "Ninth Grade Math Score" : ninth_grade_math,
                                         "Tenth Grade Math Score" : tenth_grade_math,
                                         "Eleventh Grade Math Score" : eleventh_grade_math,
                                         "Twelfth Grade Math Score" : twelfth_grade_math})


# Minor data wrangling
math_scores_by_grade.index.name = "School Name"

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# boolean mask by grade
ninth_graders = df[(df["grade"] == "9th")]
tenth_graders = df[(df["grade"] == "10th")]
eleventh_graders = df[(df["grade"] == "11th")]
twelfth_graders = df[(df["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading = ninth_graders.groupby("school_name")["reading_score"].mean()
tenth_grade_reading = tenth_graders.groupby("school_name")["reading_score"].mean()
eleventh_grade_reading = eleventh_graders.groupby("school_name")["reading_score"].mean()
twelfth_grade_reading = twelfth_graders.groupby("school_name")["reading_score"].mean()

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame ({ "Ninth Grade Reading Score" : ninth_grade_reading,
                                         "Tenth Grade Reading Score" : tenth_grade_reading,
                                         "Eleventh Grade Reading Score" : eleventh_grade_reading,
                                         "Twelfth Grade Reading Score" : twelfth_grade_reading})


# Minor data wrangling
reading_scores_by_grade.index.name = "School Name"

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Use `pd.cut` to categorize spending based on the bins.
spending_df["Spending Ranges (Per Student)"] = pd.cut(spending_df["Per Student Budget"], bins = spending_bins, labels = labels)
spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = spending_df.groupby(["Spending Ranges (Per Student)"])["Per School Math (%)"].mean()
spending_passing_reading = spending_df.groupby(["Spending Ranges (Per Student)"])["Per School Reading (%)"].mean()
overall_passing_spending = spending_df.groupby(["Spending Ranges (Per Student)"])["Passing Rate (%)"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame ({
    "Spending Math Scores" : spending_math_scores,
    "Spending Reading Scores" : spending_reading_scores,
    "Spending Passing Math" : spending_passing_math,
    "Spending Passing Reading" : spending_passing_reading,
    "Spending Overall Passing" : overall_passing_spending,
})

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Student Count"], bins = size_bins, labels = labels)
per_school_summary
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["Per School Math (%)"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["Per School Reading (%)"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Passing Rate (%)"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame ({
    "Size Math Scores" : size_math_scores,
    "Size Reading Scores" : size_reading_scores,
    "Size Passing Math" : size_passing_math,
    "Size Passing Reading" : size_passing_reading,
    "Size Overall Passing" : size_overall_passing,
})

# Display results
size_summary
```
